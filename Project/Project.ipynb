## Hypothesis testing

##### 1. What the p-value of a statistical test means and how it can be used to make conclusions.

Whenever we run statistical tests we get corresponding p-values. A statistical test compares two possible hypotheses. The p-value is used to express the statistical significance, it is the probability value and is used to either accept or reject the null hypothesis. p-value is number between 0 and 1. We compare the p-value with alpha (aka significance level). The alpha value is defined by the domain expert, if not given, it is generally considered to be 0.05. If the p-value that we derive is smaller than the alpha, it implies that it is statistically significant (indicating strong evidence against the null hypothesis) and we reject the null hypothesis. On the other hand, if p-value is larger than the alpha, we fail to reject (or accept) the null hypothesis.

##### 2. Statistical evidence to claim that the travel times tend to be shorter or longer for the single than for the season ticket type

```python
# Function to check if data is normally distributed or not
def normal_distribution(data):
    p_value = stats.shapiro(data).pvalue
    if p_value <= 0.05:
        return False
    else:
        return True
```

```python
single_ticket = DF_groupedby_ticket.get_group('single')
season_ticket = DF_groupedby_ticket.get_group('season')
savonia_ticket = DF_groupedby_ticket.get_group('savonia')
```

```python
print(normal_distribution(single_ticket.duration))
print(normal_distribution(season_ticket.duration))
```

```python
stats.mannwhitneyu(single_ticket.duration,season_ticket.duration)
```

##### **Steps**
- Check if the variables are quantitative or categorical (quantitative in this case)
- Check if the variables are normally distributed or not using Shapiro Wilk test (not normally distributed in this case)
- Next step is to compare the means and standard deviations of the sample (using Mann–Whitney U test, as the two variables are quantative, independent and not normally distributed)

**p-value of Mann–Whitney U test (0.3645632872998459) > 0.05, therefore, there is no significant difference between the travel times for the single and the season ticket type.**

##### 3. Statistical evidence to claim that the single and savonia ticket types differ with respect to how often the electric assistance is used

```python
assistance_contigency = pd.crosstab(DF.loc[(DF.ticket == 'single') | (DF.ticket == 'savonia'), 'ticket'], DF.assistance)
assistance_contigency
```

```python
stats.chi2_contingency(assistance_contigency)[1]
```

##### **Steps**
- Check the variable type (categorical in this case).
- No need to check normality, since the variables are categorical.
- Next step is to compare the two groups using chi square test, as the two variables are categorical.

 **p-value of the chi square test (4.3404376688598015e-09) < 0.05, therefore, there is statistically significant difference between electric assistance used by the two ticket types.**

##### 4. Statistical evidence to claim that the travel distances tend to be shorter or longer during one month than during the others

```python
DF_groupby_month = DF.groupby(DF.month)['distance']
```

```python
DF_groupby_month.apply(lambda grp: normal_distribution(grp))
```

```python
stats.kruskal(*[group.values for name, group in DF_groupby_month])
```

##### **Steps**
- Check the variable type (numerical in this case).
- Check if the variables are normally distributed or not using Shapiro Wilk test (not normally distributed in this case)
- Next step is to compare the difference between the sample (using Kruskal–Wallis test, as all the variables are quantative, independent and not normally distributed)

**p-value of the Kruskal–Wallis test (9.43316676101914e-10) < 0.05, therefore, there is statistically significant difference between distance travelled among all the months.**
